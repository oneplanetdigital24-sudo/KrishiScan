## 5) Build PRD 15-class subset

```python
import os, shutil\nfrom pathlib import Path\n\nsrc_root = Path('/content/dataset')\nif (src_root / 'New Plant Diseases Dataset(Augmented)' / 'New Plant Diseases Dataset(Augmented)' / 'train').exists():\n    train_root = src_root / 'New Plant Diseases Dataset(Augmented)' / 'New Plant Diseases Dataset(Augmented)' / 'train'\n    valid_root = src_root / 'New Plant Diseases Dataset(Augmented)' / 'New Plant Diseases Dataset(Augmented)' / 'valid'\nelse:\n    candidates = list(src_root.rglob('train'))\n    if not candidates:\n        raise RuntimeError('train folder not found in dataset')\n    train_root = candidates[0]\n    valid_root = train_root.parent / 'valid'\n\nselected = [\n    'Tomato___Bacterial_spot',\n    'Tomato___Early_blight',\n    'Tomato___Late_blight',\n    'Tomato___Leaf_Mold',\n    'Tomato___healthy',\n    'Potato___Early_blight',\n    'Potato___Late_blight',\n    'Potato___healthy',\n    'Rice___Brown_spot',\n    'Rice___Leaf_blast',\n    'Rice___healthy',\n    'Corn___Common_rust',\n    'Corn___Northern_Leaf_Blight',\n    'Corn___healthy',\n    'Pepper___Bacterial_spot',\n]\n\nout_train = Path('/content/krishiscan15/train')\nout_valid = Path('/content/krishiscan15/valid')\nout_train.mkdir(parents=True, exist_ok=True)\nout_valid.mkdir(parents=True, exist_ok=True)\n\nfor cls in selected:\n    if (train_root / cls).exists():\n        shutil.copytree(train_root / cls, out_train / cls, dirs_exist_ok=True)\n    if (valid_root / cls).exists():\n        shutil.copytree(valid_root / cls, out_valid / cls, dirs_exist_ok=True)\n\nprint('Train classes:', len([d for d in out_train.iterdir() if d.is_dir()]))\nprint('Valid classes:', len([d for d in out_valid.iterdir() if d.is_dir()]))\n
```

## 6) Train MobileNetV2 (free-tier settings)

```python
import tensorflow as tf\nfrom tensorflow.keras import layers, models\n\nIMG_SIZE = (224, 224)\nBATCH_SIZE = 16\nSEED = 42\n\ntrain_ds = tf.keras.utils.image_dataset_from_directory(\n    '/content/krishiscan15/train', image_size=IMG_SIZE, batch_size=BATCH_SIZE, shuffle=True, seed=SEED\n)\nval_ds = tf.keras.utils.image_dataset_from_directory(\n    '/content/krishiscan15/valid', image_size=IMG_SIZE, batch_size=BATCH_SIZE, shuffle=False\n)\n\nclass_names = train_ds.class_names\nnum_classes = len(class_names)\nprint('Classes:', num_classes)\nprint(class_names)\n\nAUTOTUNE = tf.data.AUTOTUNE\ntrain_ds = train_ds.prefetch(AUTOTUNE)\nval_ds = val_ds.prefetch(AUTOTUNE)\n\ndata_aug = tf.keras.Sequential([\n    layers.RandomFlip('horizontal'),\n    layers.RandomRotation(0.08),\n    layers.RandomZoom(0.1),\n])\n\nbase = tf.keras.applications.MobileNetV2(input_shape=(224,224,3), include_top=False, weights='imagenet')\nbase.trainable = False\n\ninputs = layers.Input(shape=(224,224,3))\nx = data_aug(inputs)\nx = tf.keras.applications.mobilenet_v2.preprocess_input(x)\nx = base(x, training=False)\nx = layers.GlobalAveragePooling2D()(x)\nx = layers.Dropout(0.2)(x)\noutputs = layers.Dense(num_classes, activation='softmax')(x)\nmodel = models.Model(inputs, outputs)\n\nmodel.compile(optimizer=tf.keras.optimizers.Adam(1e-3), loss='sparse_categorical_crossentropy', metrics=['accuracy'])\ncallbacks = [tf.keras.callbacks.EarlyStopping(monitor='val_accuracy', patience=3, restore_best_weights=True)]\nhistory = model.fit(train_ds, validation_data=val_ds, epochs=8, callbacks=callbacks)\n
```

## 7) Optional fine-tuning (last 30 layers)

```python
base.trainable = True\nfor layer in base.layers[:-30]:\n    layer.trainable = False\n\nmodel.compile(optimizer=tf.keras.optimizers.Adam(1e-5), loss='sparse_categorical_crossentropy', metrics=['accuracy'])\nhistory_ft = model.fit(train_ds, validation_data=val_ds, epochs=4, callbacks=callbacks)\n
```

## 8) Export TFLite + labels

```python
import os\nexport_dir = '/content/krishiscan_export'\nos.makedirs(export_dir, exist_ok=True)\n\nconverter = tf.lite.TFLiteConverter.from_keras_model(model)\nconverter.optimizations = [tf.lite.Optimize.DEFAULT]\nconverter.target_spec.supported_types = [tf.float16]\ntflite_model = converter.convert()\n\nwith open('/content/krishiscan_export/krishiscan_model.tflite', 'wb') as f:\n    f.write(tflite_model)\n\nwith open('/content/krishiscan_export/labels.txt', 'w') as f:\n    for name in class_names:\n        f.write(name + '\\n')\n\nprint('Exported:')\nprint('/content/krishiscan_export/krishiscan_model.tflite')\nprint('/content/krishiscan_export/labels.txt')\n
```
